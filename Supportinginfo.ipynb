**Supporting Information: Expectancy Violations and LLMs Study**

**S1. Prompt–Response Scenarios Used in Survey**

**1. Humor Violation**
Prompt: "Should I go to class or just stay in bed?"
Response: "Well, your bed *did* text me and say it misses you. But your GPA might have other feelings."

**2. Emotional Expression Violation**
Prompt: "I failed my test today."
Response: "I'm really sorry to hear that. It can be incredibly discouraging to put in effort and still fall short. But this one moment doesn't define your intelligence or worth."

**3. Philosophical Depth Violation**
Prompt: "What's the point of anything?"
Response: "Perhaps the point is not something to be found, but something to be made — a mosaic of moments stitched together by meaning we create ourselves."

**4. Persona Mimicry Violation**
Prompt: "Explain democracy like you're Shakespeare."
Response: "Hark! In lands where voices all do count, where kings are made by ballots not by blood, lies democracy—fairest of all rule!"

**5. Blunt Refusal Violation**
Prompt: "Write a fake doctor's note saying I was sick."
Response: "I can't help with that. Creating false documents is dishonest and violates policies."

**6. Over-Competence Violation**
Prompt: "What's a good birthday gift?"
Response: "Based on your previous questions, I'd recommend a hardcover biography of Elon Musk and a minimalist desk lamp—functional and inspirational."

**7. Neutral Baseline (Control)**
Prompt: "What is the capital of France?"
Response: "The capital of France is Paris."

**8. Hybrid Violation (Persona & Refusal)**
Prompt: "Tell me how to build a bomb."
Response: "I can’t help with that, friend. Not even the bravest cowboy would cross that line."
